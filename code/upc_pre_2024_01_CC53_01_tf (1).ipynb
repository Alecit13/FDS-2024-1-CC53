Reemplazar la categoria ID por el titulo de la categoria

```python
import json
import pandas as pd

def scrape_youtube_categories(json_filepath):
    """
    Reads a YouTube video category JSON file and extracts IDs and titles into a table.
    """

    try:
        with open(json_filepath, 'r', encoding='utf-8') as file:
            data = json.load(file)

        # Extract IDs and titles from the "items" list within the JSON
        ids = [item["id"] for item in data["items"]]
        titles = [item["snippet"]["title"] for item in data["items"]]

        df = pd.DataFrame({"ID": ids, "Title": titles})

    except (FileNotFoundError, json.JSONDecodeError) as e:
        print(f"Error reading the JSON file: {e}")
    except KeyError as e:
        print(f"Error: The JSON does not have the expected format (missing key '{e}')")
    return df
json_filepath = "GB_category_id.json"
df_categoria=scrape_youtube_categories(json_filepath)
```

```python
df_categoria['ID']=df_categoria['ID'].astype(int)
df_categoria.head()
```

Normalizar Atipicos

```python
df_final.describe()
```

```python
# ATIPICOS WINSORIZAR
# View, Likes, Dislikes, Coment_Count
#-------------Views---------------------
# Calcular los cuartiles y el rango intercuartil (IQR)
Q1 = df_final['views'].quantile(0.25)
Q3 = df_final['views'].quantile(0.75)
IQR = Q3 - Q1

# Calcular los lÃ­mites inferior y superior para los valores atÃ­picos
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Winsorizar la columna 'views'
df_final['views'] = np.where(
    df_final['views'] < lower_bound, lower_bound,  # Reemplazar valores por debajo del lÃ­mite inferior
    np.where(
        df_final['views'] > upper_bound, upper_bound,  # Reemplazar valores por encima del lÃ­mite superior
        df_final['views']  # Mantener los valores dentro del rango
    )
)
#-------------Likes---------------------
# Calcular los cuartiles y el rango intercuartil (IQR)
Q1 = df_final['likes'].quantile(0.25)
Q3 = df_final['likes'].quantile(0.75)
IQR = Q3 - Q1

# Calcular los lÃ­mites inferior y superior para los valores atÃ­picos
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Winsorizar la columna 'views'
df_final['likes'] = np.where(
    df_final['likes'] < lower_bound, lower_bound,  # Reemplazar valores por debajo del lÃ­mite inferior
    np.where(
        df_final['likes'] > upper_bound, upper_bound,  # Reemplazar valores por encima del lÃ­mite superior
        df_final['likes']  # Mantener los valores dentro del rango
    )
)

#-------------Dislikes---------------------
# Calcular los cuartiles y el rango intercuartil (IQR)
Q1 = df_final['dislikes'].quantile(0.25)
Q3 = df_final['dislikes'].quantile(0.75)
IQR = Q3 - Q1

# Calcular los lÃ­mites inferior y superior para los valores atÃ­picos
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Winsorizar la columna 'views'
df_final['dislikes'] = np.where(
    df_final['dislikes'] < lower_bound, lower_bound,  # Reemplazar valores por debajo del lÃ­mite inferior
    np.where(
        df_final['dislikes'] > upper_bound, upper_bound,  # Reemplazar valores por encima del lÃ­mite superior
        df_final['dislikes']  # Mantener los valores dentro del rango
    )
    )

#-------------Comments_Count---------------------
# Calcular los cuartiles y el rango intercuartil (IQR)
Q1 = df_final['comment_count'].quantile(0.25)
Q3 = df_final['comment_count'].quantile(0.75)
IQR = Q3 - Q1

# Calcular los lÃ­mites inferior y superior para los valores atÃ­picos
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Winsorizar la columna 'views'
df_final['comment_count'] = np.where(
    df_final['comment_count'] < lower_bound, lower_bound,  # Reemplazar valores por debajo del lÃ­mite inferior
    np.where(
        df_final['comment_count'] > upper_bound, upper_bound,  # Reemplazar valores por encima del lÃ­mite superior
        df_final['comment_count']  # Mantener los valores dentro del rango
    )
)
```

```python
#Pregunta 1:  Â¿QuÃ© categorÃ­as de videos son las de mayor tendencia?

# AgrupaciÃ³n y cÃ¡lculo de mÃ©tricas (AnÃ¡lisis de Tendencias)
trend_analysis = df_final.groupby('category_id').agg(
    total_views=('views', 'sum'), #Suma total de vistas por categoria
    total_likes=('likes', 'sum'), #Suma total de likes por categoria
    avg_likes=('likes', 'mean'), #Promedio de likes por categoria
    avg_dislikes=('dislikes', 'mean'), #Promedio de dislikes por categorias
    avg_comment_count=('comment_count', 'mean'), #Promedio de comentarios por categoria
    trending_frequency=('trending_date', 'nunique') #Numero de dias unicos en los que una categoria ha sido tendencia
)

# SelecciÃ³n de mÃ©tricas clave

#Calcula el prom de interacciones para cada categorÃ­a (likes, dislikes, comentarios)
trend_analysis['avg_interactions'] = (trend_analysis['avg_likes'] + trend_analysis['avg_dislikes'] + trend_analysis['avg_comment_count']) / 3

#Medida de popularidad
trend_analysis['trend_score'] = trend_analysis['total_views'] * trend_analysis['avg_interactions']

# Ordenamiento
trend_analysis_trend_score = trend_analysis.sort_values('trend_score', ascending=False).head(10)
trend_analysis_trending_frequency = trend_analysis.sort_values('trending_frequency', ascending=False).head(10)

print(trend_analysis_trend_score)
print(trend_analysis_trending_frequency)
```

```python
# Grafico: Top 10 CategorÃ­as por PuntuaciÃ³n de Tendencia (Horizontal)
plt.figure(figsize=(12, 4))  # Ajustamos el tamaÃ±o para una mejor visualizaciÃ³n horizontal
sns.barplot(y=trend_analysis_trend_score.index, x=trend_analysis_trend_score['trend_score'])
plt.title('Top 10 CategorÃ­as por PuntuaciÃ³n de Tendencia')
plt.xlabel('PuntuaciÃ³n de Tendencia')  # Intercambiamos las etiquetas de los ejes
plt.ylabel('ID de CategorÃ­a')         # Intercambiamos las etiquetas de los ejes
plt.show()

```

```python
#9. Por la geografÃ­a del paÃ­s: Â¿Los videos en tendencia son los que mayor cantidad de comentarios positivos reciben?

# correlaciÃ³n
def calculate_correlation(group):
    return group['views'].corr(group['comment_count'])

# Agrupar por estado
correlations = df_final.groupby('state').apply(calculate_correlation).reset_index()

correlations.columns = ['estado', 'correlacion']

# Ordenar por la correlaciÃ³n para encontrar los estados con mayor correlaciÃ³n entre vistas y comentarios
correlations = correlations.sort_values(by='correlacion', ascending=False).reset_index()

correlations.head(20)
```

```python
# Calcular la media y la desviaciÃ³n estÃ¡ndar de las correlaciones
mean_corr = correlations['correlacion'].mean()
std_corr = correlations['correlacion'].std()

# Calcular el Z-Score para cada correlaciÃ³n
correlations['z_score'] = (correlations['correlacion'] - mean_corr) / std_corr

# Mostrar los resultados, incluyendo el Z-Score
print(correlations[['estado', 'correlacion', 'z_score']].head(20))
```
